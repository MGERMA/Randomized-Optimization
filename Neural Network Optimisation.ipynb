```python
import mlrose
import numpy as np
import pandas as pd
from time import time
import os
from tqdm import tqdm

from itertools import product
from matplotlib import pyplot as plt
from math import *
import pandas as pd
```

```python
path = "data/list_attr_celeba.csv"
ds = pd.read_csv(path)
```

```python
#random_hill_climb
#simulated_annealing
#genetic_alg
clf = mlrose.NeuralNetwork(hidden_nodes=[20,20], activation='relu', 
              algorithm='genetic_alg', max_iters=40, bias=True, is_classifier=True, 
              learning_rate=5, early_stopping=False, clip_max=100.0, restarts=10, 
              pop_size=20, mutation_prob=0.1, max_attempts=10,
            schedule= mlrose.GeomDecay(init_temp=1.0, decay=0.99, min_temp=0.001), curve=True)


```

```python
from sklearn.model_selection import train_test_split
# Dropping NaN and other unusable values
ds = ds.dropna()
#
Y = ds.pop("Attractive")
ds.pop("image_id")
Y = [max(0,y) for y in Y]

X = ds

x_train,x_test,y_train,y_test = train_test_split(X,Y,test_size=0.3,random_state=0)

```

```python
from sklearn.neural_network import MLPClassifier

t = time()
clf = clf.fit(x_train, y_train)
elapsed = time() - t
print("Elapsed time is ",elapsed,"s")

```

```python
plt.plot(clf.fitness_curve)
plt.show()
```

```python
from sklearn.metrics import accuracy_score

accuracy_score(y_test,clf.predict(x_test))
```

```python
# Statistical Data
nb_rep = 10

#random_hill_climb

clf = mlrose.NeuralNetwork(hidden_nodes=[20,20], activation='relu', 
              algorithm='random_hill_climb', max_iters=10, bias=True, is_classifier=True, 
              learning_rate=5, early_stopping=False, clip_max=100.0, restarts=20, 
              pop_size=200, mutation_prob=0.1, max_attempts=20,
            schedule= mlrose.GeomDecay(init_temp=1.0, decay=0.99, min_temp=0.001), curve=True)

hill_score = []
for i in tqdm(range(nb_rep)) :
    t = time()
    clf = clf.fit(x_train, y_train)
    elapsed = time() - t
    hill_score.append(accuracy_score(y_test,clf.predict(x_test)))


#simulated_annealing

clf = mlrose.NeuralNetwork(hidden_nodes=[20,20], activation='relu', 
              algorithm='simulated_annealing', max_iters=10, bias=True, is_classifier=True, 
              learning_rate=5, early_stopping=False, clip_max=100.0, restarts=10, 
              pop_size=200, mutation_prob=0.1, max_attempts=10,
            schedule= mlrose.GeomDecay(init_temp=1.0, decay=0.999, min_temp=0.001), curve=True)

sa_score = []
for i in tqdm(range(nb_rep)) :
    t = time()
    clf = clf.fit(x_train, y_train)
    elapsed = time() - t
    sa_score.append(accuracy_score(y_test,clf.predict(x_test)))


#genetic_alg

clf = mlrose.NeuralNetwork(hidden_nodes=[20,20], activation='relu', 
              algorithm='genetic_alg', max_iters=10, bias=True, is_classifier=True, 
              learning_rate=5, early_stopping=False, clip_max=50.0, restarts=10, 
              pop_size=10, mutation_prob=0.1, max_attempts=10,
            schedule= mlrose.GeomDecay(init_temp=1.0, decay=0.99, min_temp=0.001), curve=True)

ga_score = []
for i in tqdm(range(nb_rep)) :
    t = time()
    clf = clf.fit(x_train, y_train)
    elapsed = time() - t
    ga_score.append(accuracy_score(y_test,clf.predict(x_test)))


```

```python
plt.rcParams["figure.figsize"] = (15,10)


plt.boxplot([hill_score, sa_score, ga_score], labels = ["Hill climbing", "Simulated Annealing", "Genetic Algorithm" ])
print(" ")
```
